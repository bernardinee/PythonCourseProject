3) Pie Chart: pH Distribution

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = r"C:\Users\Bernardine Adusei\Desktop\11123762\red_wine.csv"
df = pd.read_csv(file_path)
pH_counts = df['pH'].value_counts()

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 8))

ax1.pie(pH_counts, labels=pH_counts.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
ax1.set_title('Distribution of pH in Red Wine')
ax1.axis('equal')

table_data = pd.DataFrame({'pH Value': pH_counts.index, 'Count': pH_counts}).sort_values(by='pH Value')
table = ax2.table(cellText=table_data.values, colLabels=table_data.columns, cellLoc='center', loc='center')
table.auto_set_font_size(False)
table.set_fontsize(10)
table.scale(1.2, 1.2)  

ax2.axis('off')
plt.show()
```

9) Density Curve of pH in Red Wine

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

file_path = r"C:\Users\Bernardine Adusei\Desktop\11123762\red_wine.csv"
df = pd.read_csv(file_path)

pH_values = df['pH'].values

density, bins, _ = plt.hist(pH_values, bins=30, density=True, alpha=0.7, color='blue', edgecolor='black')

kernel_density = (1 / (density.sum() * np.diff(bins))) * np.exp(-(bins[:-1] + np.diff(bins) / 2 - pH_values[:, None])**2 / (2 * np.std(pH_values)**2))

plt.plot(bins[:-1], kernel_density.sum(axis=0), color='blue', label='Density Curve')

plt.title('Density Curve of pH in Red Wine')
plt.xlabel('pH')
plt.ylabel('Density')
plt.legend()
plt.show()
```

10) Violin Plot for pH 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

file_path = r"C:\Users\Bernardine Adusei\Desktop\11123762\red_wine.csv"
df = pd.read_csv(file_path)

plt.figure(figsize=(10, 6))
plt.violinplot(df['pH'], vert=False, showmeans=True, showextrema=True, showmedians=True)
plt.title('Violin Plot of pH in Red Wine')
plt.xlabel('pH')
plt.ylabel('Density')
plt.show()
```
